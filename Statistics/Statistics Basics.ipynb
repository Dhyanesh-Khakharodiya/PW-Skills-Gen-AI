# Statistics Basics

1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Ans:  
**Descriptive statistics** are used to summarize and describe the main features of a dataset. They provide simple numerical or visual summaries without drawing conclusions beyond the available data.  
Examples:
* Calculating the average marks of a class
* Finding the maximum, minimum, mean, median, or standard deviation
* Using charts like bar graphs or histograms to represent data
  
**Inferential statistics**, on the other hand, are used to make predictions or draw conclusions about a larger population based on a sample of data. They involve probability and hypothesis testing.
Examples:
* Predicting the average income of a city using a sample survey
* Conducting a hypothesis test to check if a new teaching method improves student performance
* Using confidence intervals to estimate population parameters

2. What is sampling in statistics? Explain the differences between random and stratified sampling.

Ans:  
Sampling in statistics is the process of selecting a subset of individuals or observations from a larger population to draw conclusions about the entire population. It saves time, reduces cost, and makes data collection practical while still allowing reliable analysis.

**Random Sampling**  
In random sampling, every member of the population has an equal and independent chance of being selected. The selection is made purely by chance.
  
Example:
Selecting 100 students randomly from a college to estimate average exam scores.

**Stratified Sampling**
  
In stratified sampling, the population is first divided into subgroups (strata) based on specific characteristics (such as age, gender, or department). Samples are then randomly selected from each stratum.
  
Example:  
Dividing students by department and then randomly selecting students from each department.

3.  Define mean, median, and mode. Explain why these measures of central tendency are important.

Ans:    
Mean, median, and mode are measures of central tendency used to describe the center or typical value of a dataset.
  
**Mean** is the average of all values, calculated by dividing the sum of the observations by the number of observations.  
It is useful when all values contribute equally and the data has no extreme outliers.
  
**Median** is the middle value when the data is arranged in ascending or descending order.  
It is important because it is not affected by extremely high or low values.  
  
**Mode** is the value that occurs most frequently in a dataset.  
It is useful for identifying the most common or popular value, especially in categorical data.  
  
**Importance of these measures**:  
They help summarize large datasets into a single representative value, make data easier to understand, allow comparison between datasets, and support better decision-making by identifying typical patterns in data.

4.  Explain skewness and kurtosis. What does a positive skew imply about the data?

Ans:   
Skewness and kurtosis are measures used to describe the shape of a data distribution.
  
**Skewness** measures the asymmetry of a distribution.
* Positive skew (right skew): The tail on the right side is longer.
* Negative skew (left skew): The tail on the left side is longer.
* Zero skew: The distribution is symmetric.
  
*What a positive skew implies:*  
A positive skew indicates that most data values are concentrated on the lower end, with a few very large values pulling the tail to the right. In such cases:
  
* $Mean > Median > Mode$
* The data contains high-value outliers
  
Example: Income distribution, where most people earn moderate incomes and a few earn extremely high incomes.
  
**Kurtosis** measures the peakedness or tail heaviness of a distribution.

* High kurtosis: Sharp peak and heavy tails (more extreme values)
* Low kurtosis: Flatter peak and light tails (fewer extreme values)

5.  Implement a Python program to compute the mean, median, and mode of a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


```python
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
```

```python
import numpy as np
from statistics import mode 
```

```python
num_mean = np.mean(numbers)
num_median = np.median(numbers)
num_mode = mode(numbers)

print(f"The mean of the given list is: {num_mean}")
print(f"The median of the given list is: {num_median}")
print(f"The mode of the given list is: {num_mode}")
```

8. You are working as a data analyst in an e-commerce company. Themarketing team wants to know if there is a relationship between advertising spend and daily sales.
* Explain how you would use covariance and correlation to explore this relationship.
* Write Python code to compute the correlation between the two lists:  
advertising_spend = [200, 250, 300, 400, 500]  
daily_sales = [2200, 2450, 2750, 3200, 4000]


```python
import pandas as pd

ecom_df = pd.DataFrame({
    "advertising_spend" : [200, 250, 300, 400, 500],
    "daily_sales" : [2200, 2450, 2750, 3200, 4000]
})

ecom_df.head()
```

```python
df_cov = ecom_df.cov()
df_corr = ecom_df.corr()
```

```python
df_cov
```

```python
df_corr
```

To understand whether advertising spend is related to daily sales, I would use both covariance and correlation:
  
**Covariance** indicates the direction of the relationship.

* A positive covariance means that as advertising spend increases, daily sales also tend to increase.
* A negative covariance would indicate an inverse relationship.
* However, covariance does not indicate the strength of the relationship and is affected by the scale of the data.
  
**Correlation** measures both the strength and direction of the relationship on a standardized scale from 1 to +1.
  
* A value close to +1 indicates a strong positive relationship.
* A value close to 0 indicates little or no linear relationship.
* This makes correlation more interpretable and suitable for business analysis.
  
In practice, I would rely more on correlation for decision-making, while using covariance as a preliminary directional check.

Here the Correlation indicates that there is a very strong postivie relationship between advertising spend and daily sales

9. Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
* Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) youd use.
* Write Python code to create a histogram using Matplotlib for the survey data:  
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]


Ans:  
To understand the distribution of customer satisfaction scores (110) before launching a new product, I would use the following:
  
* Mean  to understand the average satisfaction level of customers.
* Median  to identify the central value and check if the data is skewed.
* Standard deviation  to measure how spread out the satisfaction scores are from the average.
* Minimum and maximum  to see the range of customer responses.
  
For visualization:
  
A histogram is most suitable because it shows how frequently each score occurs, helping identify patterns such as clustering, skewness, or gaps in satisfaction levels.

```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=7)
plt.xlabel("Customer Satisfaction Score")
plt.ylabel("Frequency")
plt.title("Distribution of Customer Satisfaction Scores")
plt.show()
```

Interpretation
  
The histogram visually displays how customer satisfaction scores are distributed across the scale. It helps identify whether most customers are satisfied, whether responses are evenly spread, or if there are any extreme values. This insight is valuable for assessing readiness before a new product launch.
