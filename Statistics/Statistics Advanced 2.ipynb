1. What is hypothesis testing in statistics?

Ans:  
**Hypothesis testing** in statistics is a method used to make decisions or draw conclusions about a population based on sample data. It involves proposing an initial assumption, called the null hypothesis, and then using statistical tests to determine whether there is enough evidence to reject that assumption.  
  
Typically, two hypotheses are defined:
  
* Null hypothesis (H‚ÇÄ): assumes no effect or no difference

* Alternative hypothesis (H‚ÇÅ): suggests that an effect or difference exists
  
Based on the sample data and a chosen significance level, a statistical test determines whether the observed results are likely due to chance or indicate a meaningful relationship.

2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?


Ans:  
The null hypothesis (H‚ÇÄ) is the initial assumption that there is no effect, no difference, or no relationship between variables. It represents the default position that any observed change is due to chance.  
  
The alternative hypothesis (H‚ÇÅ or Ha) is the opposite claim, suggesting that a real effect, difference, or relationship exists in the data.
  
Key differences:  
  
* Null hypothesis: assumes nothing has changed or no association exists.
* Alternative hypothesis: proposes that a significant change or association is present.
  
During hypothesis testing, statistical evidence is used to decide whether to reject the null hypothesis in favor of the alternative.

3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.


Ans:  
The significance level, usually denoted by `ùõº`, is the threshold used to decide whether the result of a hypothesis test is statistically meaningful. It represents the probability of rejecting the null hypothesis when it is actually true (Type I error).  
  
Common significance levels are 0.05, 0.01, or 0.10, with 0.05 being widely used in practice.
  
Role in deciding the outcome:
  
* After performing a statistical test, a p-value is obtained. 
* If the p-value ‚â§ significance level, the null hypothesis is rejected because the result is considered statistically significant.
* If the p-value > significance level, there is not enough evidence to reject the null hypothesis.
  
In simple terms, the significance level sets the decision boundary for determining whether the observed results are likely due to chance or reflect a real effect.

4. What are Type I and Type II errors? Give examples of each

Ans:  
Type I and Type II errors are mistakes that can occur when making decisions in hypothesis testing.
  
**Type I Error (False Positive)**:  
  
A Type I error happens when the null hypothesis is rejected even though it is actually true. In simple terms, it means detecting an effect that does not really exist.
  
Example:  
An e-commerce company concludes that a new advertising campaign increases sales, but in reality, the campaign has no real impact. The company spends more on marketing based on a false result.
  
**Type II Error (False Negative)**:
  
A Type II error occurs when the null hypothesis is not rejected even though it is false. This means failing to detect a real effect.
  
Example:  
A company tests a new product feature that actually improves customer satisfaction, but the test results suggest no improvement. As a result, the company decides not to launch a beneficial feature.

5. What is the difference between a Z-test and a T-test? Explain when to use each.

Ans:  
The main difference between a Z-test and a T-test lies in the sample size and whether the population standard deviation is known.
  
**Z-test:**    
A Z-test is used when:  
  
* The population standard deviation is known, or
* The sample size is large (typically `ùëõ ‚â• 30`)
* The data is approximately normally distributed.
  
Example:  
Testing whether the average delivery time of a large number of orders differs from a known company benchmark.
  
**T-test:**   
A T-test is used when:  
  
* The population standard deviation is unknown
* The sample size is small (usually `ùëõ<30`)
* The data is approximately normally distributed.
  
Example:  
Comparing the average satisfaction score of a small group of customers before and after a feature change.

8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulating data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=200)  # mean=50, std=5, sample size=200

# Calculating 95% confidence interval for the mean
mean_value = np.mean(data)
std_error = stats.sem(data)  # standard error of mean
confidence_interval = stats.t.interval(
    confidence=0.95,
    df=len(data)-1,
    loc=mean_value,
    scale=std_error
)

print("Sample Mean:", mean_value)
print("95% Confidence Interval:", confidence_interval)

# Plotting histogram using Matplotlib
plt.figure()
plt.hist(data, bins=20)
plt.title("Histogram of Simulated Normal Distribution Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset from seaborn
df = sns.load_dataset("tips")

# Use a numeric column (total_bill)
data = df["total_bill"].dropna().values
```

```python
# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("First 10 Z-scores:", z_scores[:10])
```

```python
# Plot histogram of standardized data
plt.figure()
plt.hist(z_scores, bins=20)
plt.title("Histogram of Z-score Standardized Data (Seaborn Tips Dataset)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```

A Z-score tells how many standard deviations a value is from the mean:

Z = 0 ‚Üí exactly at the mean

Positive Z-score ‚Üí above average

Negative Z-score ‚Üí below average

|Z| > 2 or 3 ‚Üí potentially unusual or outlier values

After standardization, the data becomes centered around 0 with a standard deviation of 1, which is why the histogram appears balanced around zero
