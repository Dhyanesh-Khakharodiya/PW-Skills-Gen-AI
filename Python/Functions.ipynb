1. What is the difference between a function and a method in Python?

Ans:  
In Python, a function is a block of reusable code that is defined independently and can be called directly by its name. It operates on the data passed to it as arguments.  
  
A method is also a function, but it is associated with an object and is called using that object. Methods usually act on the data contained within the object.  
  
In short:  
Function → independent, called directly  
Method → belongs to an object, called using the object

2. Explain the concept of function arguments and parameters in Python.

Ans:  
In Python, parameters are the variables defined in a function’s definition that specify the data the function expects to receive. Arguments are the actual values passed to the function when it is called.  
  
In simple terms, parameters act as placeholders, while arguments provide real data during function execution.

3. What are the different ways to define and call a function in Python?

Ans:  
In Python, a function can be defined and called in several ways depending on how arguments are passed. The main ways are:  
  
A function can be defined without parameters and called directly when no input is required.

```python
def greet():
    print("Hello")

greet()
```

A function can be defined with parameters and called by passing values as arguments.

```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

Functions can also be called using keyword arguments, where values are passed by specifying parameter names.

```python
def display(name, age):
    print(name, age)

display(age=20, name="Amit")
```

Default arguments allow parameters to have predefined values, which are used if no argument is provided.

```python
def welcome(name="Guest"):
    print("Welcome", name)

welcome()
welcome("Dhyan")
```

4. What is the purpose of the `return` statement in a Python function?

Ans:  
The purpose of the return statement in a Python function is to send a result back to the caller and end the execution of the function. It allows the function to pass computed values to other parts of the program. If no return statement is used, the function returns None by default.

5. What are iterators in Python and how do they differ from iterables?

Ans:  
In Python, iterables are objects that can be looped over, such as lists, tuples, strings, and dictionaries. They can produce an iterator when required.  
Iterators are objects that actually perform the iteration. They keep track of the current position and return one element at a time using the next() function until all elements are exhausted.
  
In short:  
Iterable → can be iterated over  
Iterator → performs the iteration and returns values one by one

6. Explain the concept of generators in Python and how they are defined.

Ans:  
Generators in Python are a special type of function used to produce values one at a time, instead of returning all values at once. They help save memory by generating values only when needed.

Generators are defined like normal functions but use the yield keyword instead of return. Each time yield is executed, the function pauses and remembers its state, resuming from that point the next time it is called.

7.  What are the advantages of using generators over regular functions?

Ans:  
Generators offer several advantages over regular functions:  

* They are memory efficient because values are generated one at a time instead of being stored all at once.
* They allow lazy evaluation, meaning values are produced only when needed.
* They are well suited for handling large datasets or streams of data.
* They improve performance when working with sequences that do not need to be fully stored in memory.

8. What is a lambda function in Python and when is it typically used?

Ans:  
A lambda function in Python is a small, anonymous function defined using the lambda keyword. It can have any number of arguments but only one expression.
  
Lambda functions are typically used for short, simple operations where defining a full function would be unnecessary, such as inside functions like map(), filter(), or sorted() for concise and readable code.

9. Explain the purpose and usage of the `map()` function in Python.

Ans:  
The map() function is used to apply a given function to each element of an iterable and return the results as a new iterable. It helps perform transformations on data without using explicit loops.  
  
The function takes two main arguments: a function and an iterable. The function is applied to every element of the iterable, and the transformed values are returned in order.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans:  
The functions map(), filter(), and reduce() are used for processing iterables, but they serve different purposes.  
  
* map() applies a function to each element of an iterable and returns the transformed results.

* filter() selects elements from an iterable that satisfy a given condition.

* reduce() applies a function cumulatively to the elements of an iterable to produce a single result.
  
Together, they support functional-style programming and help write concise and readable code.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even(l):
    total = 0
    for i in l:
        if i%2 == 0:
            total+=i
    return total
```

```python
list_1 = [1,2,3,4,5,6]
print(sum_of_even(list_1))
```

```python
print(sum_of_even(list(range(1,21))))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200

```python
def print_primes():
    for num in range(2, 201):
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num)
```

```python
print_primes()
```

```python
## This function checks if the number is prime or not

def is_prime(num):
    if num <= 1:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
```

```python
is_prime(13)
```

```python
is_prime(197)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
def fibonnaci(terms):
    a,b = 0,1
    count = 0

    while count < terms:
        yield a
        a,b = b,a+b
        count+=1
```

```python
for i in fibonnaci(6):
    print(i)
```

```python
fib_5 = fibonnaci(5)
```

 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
```

```python
for value in powers_of_two(5):
    print(value)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_line_by_line(file_name):
  with open(file_name,"r",encoding  = "utf-8") as file:
    for line in file:
      yield line
```

```python
for line in read_line_by_line("test.txt"):
  print(line)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python

def remove_vowels(input_string):
#input_string = "Hello World!"
  vowels = "aeiouAEIOU"
  return "".join(list(filter(lambda ch: ch not in vowels,input_string)))


print(remove_vowels("The World is beautiful!"))
```

11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:  
  
Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.
  

Write a Python program using lambda and map.

```python
# List of orders: [order_number, price_per_item, quantity]
orders = [
    [34587, 40.95, 4],
    [98762, 56.80, 5],
    [77226, 32.92, 3],
    [88112, 24.99, 3]
]

result = list(
    map(
        lambda o: (
            o[0],
            o[1] * o[2] + 10 if o[1] * o[2] < 100 else o[1] * o[2]
        ),
        orders
    )
)

print(result)
```
